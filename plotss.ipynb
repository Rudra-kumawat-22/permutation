```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
d=pd.read_csv("G:\\kittu_data\\final.csv")
```

```python
d.head()
```

```python
plt.figure(figsize=[10,5])
plt.title("departure date vs price for all flights")
sns.lineplot(d['Dept_date'],d['Price'],hue=d['Dept_month'])
plt.savefig("price vs date for all flights.png")
```

```python
sns.boxplot(d['Price'])
```

```python
d.columns
```

```python
plt.figure(figsize=[11,8])
plt.subplot(2, 1, 1)
plt.title("cabins booking counts // prices for each cabin")
sns.countplot(d['Cabin'])
plt.subplot(2, 1, 2)
sns.barplot(d['Cabin'],d['Price'])
plt.savefig('cabin_price.png')
```

```python
d.head()
```

```python

plt.figure(figsize=[11,8])
plt.subplot(2, 1, 1)
plt.title("TOTAL FLIGHTS AT EACH STOP AND VARIATION IN PRICE DUE TO STOPS")
sns.countplot(d['stops'])
plt.subplot(2, 1, 2)
sns.barplot(d['stops'],d['Price'])
plt.savefig('stops.png')
```

```python
plt.figure(figsize=[11,5])
plt.subplot(2, 1, 1)
plt.title("flights departing at different time and price variation due to differnt time ")
sns.countplot(d['Dept_flights_time'])
plt.subplot(2, 1, 2)
sns.barplot(d['Dept_flights_time'],d['Price'])
plt.savefig('flight_time.png')
```

```python
d.head()
```

```python
sns.lineplot(d['duration'],d['Price'])
plt.title("DURATION VS PRICE")
plt.savefig('duration vs price.png')
```

```python
plt.title("variation of price due to duration for each columns")
sns.lineplot(d['duration'],d['Price'],hue=d['Cabin'])
plt.savefig('duration vs price for each cabin.png')
```

```python
d.head()
```

```python
plt.figure(figsize=[10,10])
plt.title("variation of price w.r.t date for each cabin")
sns.lineplot(d['Dept_date'],d['Price'],hue=d['Cabin'])
plt.savefig('price_vs_date.png')
```
